```python
import numpy as np
n = 20
# %02d ensures that numbers have two place values.
labels = ['age']+['%02d'%i for i in range (1, n+1)]
l = [100] + [x for x in np.random.rand(n)*10+20]
with open('heights_sequoias1.csv', 'w') as fout:
    print(','.join(str(x) for x in labels), file = fout)
    print(','.join(str(x) for x in l), file = fout)
    for age in range(200, 2901, 100):
        # We don't want the trees to shrink but to grow.
        newl = [age] + [4*np.random.rand(1)[0]+x for x in l[1:]]
        print(','.join(str(l) for l in newl), file = fout)
        l = newl
```
